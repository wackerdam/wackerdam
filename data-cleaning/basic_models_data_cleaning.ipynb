```python
#q2
inpDf = inpDf.dropna()
inpDf
```

```python
#q3
#The level of measurement of the division variable is nominal. Nominal variables can be converted to numeric values by 
#assigning a numerical value to each nominal value. e.g, sales = 0, support = 1
#It is called the "The Nominal to Numerical operator"
import warnings
warnings.filterwarnings('ignore')
inpDf["division"] = inpDf["division"].astype('category')
inpDf["division"] = inpDf["division"].cat.codes

```

```python
#q4
#The level of measurement for "wage" is ordinal. Ordinal variables can be converted by grouping the observations together
#e.g values with low can be grouped and assigned 0, medium can be grouped and assigned 1, etc.
#This is called the "Ordinal to Numerical operator"
inpDf["wage"] = inpDf["wage"].astype('category')
inpDf["wage"] = inpDf["wage"].cat.codes
print
```

```python
#q7
#Why should the "avgMtlyHrs" variable be standardised: The "avgMtlyHrs" variable should be standardized
#so that we can consistently exchange data between different systems

inpDf["avgMtlyHrs"] = (inpDf["avgMtlyHrs"] - inpDf["avgMtlyHrs"].mean())/inpDf["avgMtlyHrs"].std()
print(inpDf["avgMtlyHrs"])
```

```python
#q8
#Differences between supervised and unsupervised learning:
#The difference between supervised vs unsupervised learning is the need for labelled training data. 
#Supervised machine learning relies on labelled input and output training data,
#whereas unsupervised learning processes unlabelled or raw data.Another difference is that unsupervised learning only has
#one axis or dimension whereas supervised learning has two dimensions

yDf = inpDf["resigned"]
xDf = inpDf.drop(["resigned"],axis=1)
```

```python
#q9
#Explain which type of supervised learning should be preferred, knowing that the label is the "resigned" variable:
#the "resigned" label is a binary label therefore we should use the supervised binary classification or binary classifier.

```

```python
#q10
#Explain the importance of splitting the sample into train and test sets and clarify
#which set typically receives the larger quantity of examples: The train set is used to train the model, 
#while the test set is used to evaluate the model. The train set should be larger than the test set.

from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(xDf,yDf,test_size=0.3,random_state=0)

```

```python
#q11
#Provide a brief definition of the Support Vector Machine(SVM):
#SVM is a supervised machine learning algorithm used for classification and regression tasks. 
#It works by finding the optimal hyperplane that separates different classes in the input feature space.
#The hyperplane is chosen such that it maximizes the margin, which is the distance between the hyperplane
#and the nearest data points (support vectors) from each class.

from sklearn.svm import SVC
svc = SVC(C=100,kernel="rbf",gamma=1,random_state=0)

svc.fit(xTrain,yTrain)
yPred = svc.predict(xTest)
from sklearn.metrics import accuracy_score
print(accuracy_score(yTest,yPred))
```

```python
#q12
#Provide a brief definition of Random Forests:
#Supervised learning models for classification and regression tasks RF builds multiple decision trees and merges them
#together to get a more accurate and stable prediction.

from sklearn.ensemble import RandomForestClassifier
for i in [2,10,50]:
    rfc = RandomForestClassifier(n_estimators=i,random_state=0)
    rfc.fit(xTrain,yTrain)
    yPred = rfc.predict(xTest)
    print(accuracy_score(yTest,yPred))
```
